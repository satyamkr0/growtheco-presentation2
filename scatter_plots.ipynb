```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1=pd.read_csv('data_2 - real gdp per capita.csv')
df1.head()
```

```python
df2=pd.read_csv('data_2 - 1980.csv')
df2.head()
```

```python
df3=pd.merge(df1,df2,on='Country')
df3.head()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,8))
sns.regplot(y='Average Rate of growth of real gdp per capita', x='ln(per capita real gdp)', data=df3, ci=None)

plt.xlabel('ln(Real GDP per capita) for 1980',fontsize=14)
plt.ylabel('Average growth rates',fontsize=14)
plt.title('156 Countries',fontsize=17)
```
